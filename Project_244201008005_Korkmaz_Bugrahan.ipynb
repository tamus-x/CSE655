```python
# -*- coding: utf-8 -*-
"""
Data Preprocessing & Feature Engineering for a Real-Time PI-LSTM Model.

This script implements the full data processing pipeline for our final strategy,
where the model learns from the sequence of samples WITHIN a single cycle.

The pipeline performs the following steps for each battery cell data file:
1.  Loads the raw data from Excel files.
2.  Performs a detailed data health check and reports its findings.
3.  Imputes missing values where necessary, with visual verification.
4.  Conducts Exploratory Data Analysis (EDA) on the raw data.
5.  Engineers new, powerful features, including the dynamic 'Interpolated_SoH'.
6.  Combines all processed data into a single master file, ready for model training.

The script is structured using classes to adhere to SOLID principles and is
designed to be a "glassy box" with transparent logging and visualization.
"""
# NOTE: To read .xlsx files, pandas requires the 'openpyxl' engine.
# In a Colab environment, you may need to install it first by running:
# !pip install -q openpyxl

import os
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from typing import List, Dict, Tuple

# --- Configuration ---
class Config:
    """A class to hold all configuration parameters."""
    # Input/Output paths in Google Drive
    BASE_DATA_PATH: str = "/content/drive/MyDrive/Battery2/"
    RAW_DATA_PATH: str = os.path.join(BASE_DATA_PATH, "raw_data/")
    PROCESSED_DATA_PATH: str = os.path.join(BASE_DATA_PATH, "processed_data_sample_level/")

    # EDA parameters
    PLOT_PALETTE: str = "pastel"
    HEATMAP_CMAP: str = "YlGnBu"

class DataLogger:
    """A simple class for styled and verbose logging."""
    @staticmethod
    def info(message: str) -> None:
        """Prints an informational message."""
        print(f"‚úÖ [INFO] {message}")

    @staticmethod
    def step(step_number: int, message: str) -> None:
        """Prints a step in the process."""
        print(f"\n{'='*50}\n‚ñ∂Ô∏è  STEP {step_number}: {message}\n{'='*50}")

    @staticmethod
    def warning(message: str) -> None:
        """Prints a warning message."""
        print(f"‚ö†Ô∏è [WARNING] {message}")

    @staticmethod
    def dataframe(df: pd.DataFrame, title: str) -> None:
        """Prints the head of a dataframe with a title."""
        print(f"\n--- {title} ---\n")
        print(df.head())
        print("\n")

class BatteryDataProcessor:
    """
    Encapsulates the processing logic for a single battery cell file.
    Follows the Single Responsibility Principle: its only job is to process
    one battery's data from raw to feature-engineered.
    """

    def __init__(self, file_path: str, config: Config):
        """Initializes the processor for a given file."""
        self.file_path: str = file_path
        self.config: Config = config
        self.cell_name: str = os.path.basename(file_path).split('.')[0]
        self.raw_df: pd.DataFrame = None
        self.processed_df: pd.DataFrame = None
        DataLogger.info(f"Initialized processor for battery cell: {self.cell_name}")

    def execute_pipeline(self) -> pd.DataFrame:
        """Executes the full preprocessing and feature engineering pipeline."""
        self._load_data()
        self._run_health_check_and_impute()
        self._perform_exploratory_data_analysis()
        self._engineer_features() # This is the corrected method
        return self.processed_df

    def _load_data(self):
        """Loads data from the specified Excel file."""
        DataLogger.info(f"Loading raw data from: {self.file_path}")
        self.raw_df = pd.read_excel(self.file_path, engine='openpyxl')
        DataLogger.dataframe(self.raw_df, f"Raw Data Snippet for {self.cell_name}")

    def _run_health_check_and_impute(self):
        """
        Performs data health checks and imputation with detailed reporting.
        """
        DataLogger.info("Running Data Health Check and Imputation...")
        df = self.raw_df.copy()

        # Report on missing values (assuming 0 represents a missing measurement)
        original_zeros = df[df['Internal_Resistance(Ohm)'] == 0].shape[0]
        total_rows = df.shape[0]
        missing_percentage = (original_zeros / total_rows) * 100
        print(f"   - Found {original_zeros} of {total_rows} entries ({missing_percentage:.2f}%) with 0 resistance.")

        # Impute Internal Resistance using forward-fill then backward-fill
        df['Internal_Resistance_Original'] = df['Internal_Resistance(Ohm)'].replace(0, np.nan)
        df['Internal_Resistance(Ohm)'] = df['Internal_Resistance(Ohm)'].replace(0, np.nan).ffill().bfill()

        # Visual Verification of Imputation
        plt.figure(figsize=(12, 6))
        sample_size = min(500, total_rows)
        plt.plot(df.index[:sample_size], df['Internal_Resistance_Original'][:sample_size],
                 'o', markersize=4, label='Original Measurements (where > 0)', color=sns.color_palette(self.config.PLOT_PALETTE)[1])
        plt.plot(df.index[:sample_size], df['Internal_Resistance(Ohm)'][:sample_size],
                 '-', label='Imputed Resistance', color=sns.color_palette(self.config.PLOT_PALETTE)[0], linewidth=2, zorder=-1)
        plt.title(f'Imputation of Internal Resistance for {self.cell_name}')
        plt.xlabel('Data Point Index')
        plt.ylabel('Internal Resistance (Ohm)')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.show()

        final_missing = df['Internal_Resistance(Ohm)'].isna().sum()
        if final_missing == 0:
            DataLogger.info("Imputation successful. No missing resistance values remain.")
        else:
            DataLogger.warning(f"{final_missing} missing values remain after imputation.")

        df.drop(columns=['Internal_Resistance_Original'], inplace=True)
        self.processed_df = df

    def _perform_exploratory_data_analysis(self):
        """Performs and displays EDA plots and stats on the raw (but imputed) data."""
        DataLogger.info(f"Performing Exploratory Data Analysis for {self.cell_name}...")
        sns.set_palette(self.config.PLOT_PALETTE)

        # Descriptive Statistics Table
        stats = self.processed_df[['Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)']].describe()
        DataLogger.dataframe(stats, "Descriptive Statistics (Sample-Level)")

        # Raw Sample-Level Correlation Matrix
        DataLogger.info(f"Plotting correlation matrix for raw sample-level data...")
        plt.figure(figsize=(10, 8))
        raw_corr_cols = ['Cycle_Index', 'Current(A)', 'Voltage(V)', 'Internal_Resistance(Ohm)', 'Discharge_Capacity(Ah)', 'Charge_Capacity(Ah)']
        raw_corr = self.processed_df[raw_corr_cols].corr()
        sns.heatmap(raw_corr, annot=True, fmt=".2f", cmap=self.config.HEATMAP_CMAP, linewidths=.5)
        plt.title(f'Raw Sample-Level Feature Correlation for {self.cell_name}')
        plt.show()

    def _engineer_features(self):
        """
        Engineers new features, including a physically meaningful
        'Interpolated_SoH' that degrades only during discharge.
        """
        DataLogger.info(f"Engineering features for {self.cell_name}...")
        df = self.processed_df.copy()

        # --- Step 1: Calculate Cycle-Level Summary Information ---
        # Group data by cycle to get end-of-cycle values.
        # We use 'Discharge_Capacity(Ah)' as it represents the cumulative discharge over the battery's life.
        cycle_summary = df.groupby('Cycle_Index').agg(
            Cycle_End_Cumulative_Discharge=('Discharge_Capacity(Ah)', 'max')
        ).reset_index()

        # Calculate the capacity discharged in each specific cycle.
        cycle_summary['Per_Cycle_Capacity'] = cycle_summary['Cycle_End_Cumulative_Discharge'].diff().fillna(cycle_summary['Cycle_End_Cumulative_Discharge'])
        # The baseline capacity is from the first full discharge cycle.
        initial_capacity = cycle_summary.loc[0, 'Per_Cycle_Capacity']
        if initial_capacity == 0:
            DataLogger.warning("Initial cycle capacity is 0. SoH will be NaN or infinite. Finding next valid cycle.")
            initial_capacity = cycle_summary[cycle_summary['Per_Cycle_Capacity'] > 0]['Per_Cycle_Capacity'].iloc[0]

        # Calculate SoH at the END of each cycle.
        cycle_summary['SoH_end_of_cycle'] = (cycle_summary['Per_Cycle_Capacity'] / initial_capacity) * 100

        # Calculate SoH at the START of each cycle (which is the end of the previous one).
        cycle_summary['SoH_start_of_cycle'] = cycle_summary['SoH_end_of_cycle'].shift(1).fillna(100.0)

        # Calculate the cumulative discharge capacity at the end of the PREVIOUS cycle.
        # This is the correct baseline for calculating in-cycle progress.
        cycle_summary['Prev_Cycle_End_Cumulative_Discharge'] = cycle_summary['Cycle_End_Cumulative_Discharge'].shift(1).fillna(0)

        # --- Step 2: Merge Cycle-Level Info back to Sample-Level Dataframe ---
        # Select only the new columns to merge to avoid overwriting original data.
        merge_cols = ['Cycle_Index', 'SoH_start_of_cycle', 'SoH_end_of_cycle', 'Per_Cycle_Capacity', 'Prev_Cycle_End_Cumulative_Discharge']
        df = pd.merge(df, cycle_summary[merge_cols], on='Cycle_Index', how='left')

        # --- Step 3: Create the Sample-Level Interpolated SoH Feature ---
        # Identify discharge steps (Current < 0).
        is_discharging = df['Current(A)'] < 0

        # Calculate discharge progress within the current cycle for each sample.
        # This is the current sample's total discharge minus the total from all previous cycles.
        df['Discharge_Progress_In_Cycle'] = (df['Discharge_Capacity(Ah)'] - df['Prev_Cycle_End_Cumulative_Discharge'])

        # Calculate the fraction of the cycle that has been discharged.
        # Use np.where to avoid division by zero if a cycle has no discharge.
        df['Discharge_Fraction'] = np.where(df['Per_Cycle_Capacity'] > 0,
                                            df['Discharge_Progress_In_Cycle'] / df['Per_Cycle_Capacity'],
                                            0)
        df['Discharge_Fraction'] = df['Discharge_Fraction'].clip(0, 1) # Ensure fraction is between 0 and 1.

        # Calculate the total SoH drop that occurs over the entire cycle.
        soh_drop_per_cycle = df['SoH_start_of_cycle'] - df['SoH_end_of_cycle']

        # Interpolate SoH. It only changes during discharge.
        # For charge/rest, it holds the value from the start of the cycle.
        df['Interpolated_SoH'] = df['SoH_start_of_cycle'] - (soh_drop_per_cycle * df['Discharge_Fraction'])
        df['Interpolated_SoH'] = np.where(is_discharging, df['Interpolated_SoH'], df['SoH_start_of_cycle'])

        DataLogger.info("Feature engineering complete. 'Interpolated_SoH' created.")
        DataLogger.dataframe(df[['Cycle_Index', 'Voltage(V)', 'Current(A)', 'Interpolated_SoH', 'Discharge_Fraction']], "Final Dataframe Snippet with New Features")

        # Clean up intermediate columns before finalizing
        df.drop(columns=['Per_Cycle_Capacity', 'Prev_Cycle_End_Cumulative_Discharge',
                         'Discharge_Progress_In_Cycle', 'Discharge_Fraction'], inplace=True)

        self.processed_df = df


class ProcessingPipeline:
    """Orchestrates the entire data processing workflow for multiple battery cells."""
    def __init__(self, config: Config):
        self.config = config
        self.all_cell_data = []

    def run(self, reduce_features: bool = False, features_to_keep: List[str] = None):
        """
        Runs the full pipeline from loading to saving.
        Optionally reduces features in the final output.
        """
        DataLogger.step(1, "Starting Data Processing Pipeline")
        self._create_output_directory()

        file_paths = glob.glob(os.path.join(self.config.RAW_DATA_PATH, "*.xlsx"))
        if not file_paths:
            DataLogger.warning(f"No Excel (.xlsx) files found in {self.config.RAW_DATA_PATH}. Please check the path.")
            return
        DataLogger.info(f"Found {len(file_paths)} battery data files to process.")

        for path in file_paths:
            try:
                processor = BatteryDataProcessor(path, self.config)
                processed_df = processor.execute_pipeline()
                processed_df['cell_id'] = processor.cell_name
                self.all_cell_data.append(processed_df)
                DataLogger.info(f"Successfully processed {processor.cell_name}.")
            except Exception as e:
                DataLogger.warning(f"Failed to process {path}. Error: {e}")

        if not self.all_cell_data:
            DataLogger.warning("No data was processed. Exiting pipeline.")
            return

        DataLogger.step(2, "Combining and Saving Processed Data")
        self._save_data(reduce_features, features_to_keep)
        DataLogger.step(3, "Pipeline Finished Successfully!")

    def _create_output_directory(self):
        """Creates the output directory if it doesn't exist."""
        os.makedirs(self.config.PROCESSED_DATA_PATH, exist_ok=True)
        DataLogger.info(f"Ensured output directory exists: {self.config.PROCESSED_DATA_PATH}")

    def _save_data(self, reduce_features: bool, features_to_keep: List[str]):
        """Combines all processed data and saves to a single file."""
        if not self.all_cell_data:
            return

        combined_df = pd.concat(self.all_cell_data, ignore_index=True)

        # Save the full, feature-engineered dataset
        full_output_path = os.path.join(self.config.PROCESSED_DATA_PATH, "all_cells_full_features.csv")
        combined_df.to_csv(full_output_path, index=False)
        DataLogger.info(f"Full data for {len(self.all_cell_data)} cells saved to: {full_output_path}")
        DataLogger.dataframe(combined_df, "Full Saved Data Snippet")

        # Optionally save a reduced feature set
        if reduce_features:
            if features_to_keep is None:
                # Default minimal feature set for the model
                features_to_keep = ['cell_id', 'Cycle_Index', 'Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)', 'Interpolated_SoH']

            # Ensure essential columns are present
            for col in ['cell_id', 'Cycle_Index', 'Interpolated_SoH']:
                if col not in features_to_keep:
                    features_to_keep.insert(0, col)

            reduced_df = combined_df[features_to_keep]
            reduced_output_path = os.path.join(self.config.PROCESSED_DATA_PATH, "all_cells_reduced_features.csv")
            reduced_df.to_csv(reduced_output_path, index=False)
            DataLogger.info(f"Reduced feature set saved to: {reduced_output_path}")
            DataLogger.dataframe(reduced_df, "Reduced Feature Set Snippet")


def main():
    """Main function to run the data processing pipeline."""
    # This block is for Google Colab integration.
    # If running locally, you can comment this out and ensure your paths are correct.
    try:
        from google.colab import drive
        print("üöÄ Mounting Google Drive...")
        drive.mount('/content/drive', force_remount=True)
        print("‚úÖ Google Drive mounted successfully.")
    except ImportError:
        DataLogger.warning("Not running in Google Colab. Make sure your data paths in Config are set correctly for a local environment.")
        # If running locally, you might need to adjust the BASE_DATA_PATH
        # For example: Config.BASE_DATA_PATH = "./"

    print("\n--- Battery Data Preprocessing and Feature Engineering ---")

    pipeline_config = Config()
    pipeline = ProcessingPipeline(pipeline_config)

    # --- USER CONTROL ---
    # Set to True to create a smaller file with only specified features
    PERFORM_FEATURE_REDUCTION = True

    # Define the features you want in the final reduced file.
    # If None, a default set will be used.
    # These are your model inputs (X) and target (y).
    MODEL_FEATURES = [
        'Voltage(V)',
        'Current(A)',
        'Internal_Resistance(Ohm)',
        'Step_Time(s)',
        'Interpolated_SoH' # This is the target variable (y)
    ]

    pipeline.run(reduce_features=PERFORM_FEATURE_REDUCTION, features_to_keep=MODEL_FEATURES)


if __name__ == '__main__':
    main()
```

```python
# -*- coding: utf-8 -*-
"""
Part 2: Unified Stacked Physics-Informed LSTM (PI-LSTM) for SoH Prediction.

This script builds, trains, and evaluates a single, unified stacked PI-LSTM model
using the entire 'all_cells_reduced_features.csv' dataset. This version does not
split data by cell_id for training but evaluates the final model's performance
on each cell individually.
"""

import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import kerastuner as kt
from typing import List, Dict, Tuple, Any

# --- Configuration ---
class Config:
    """A class to hold all configuration parameters for the PI-LSTM model."""
    BASE_PATH: str = "/content/drive/MyDrive/Battery2/"
    DATA_PATH: str = os.path.join(BASE_PATH, "processed_data_sample_level/")
    MODEL_PATH: str = os.path.join(BASE_PATH, "models/")
    REDUCED_DATA_FILE: str = "all_cells_reduced_features.csv"
    UNIFIED_MODEL_NAME: str = "unified_pi_lstm_model.h5"
    FEATURES: List[str] = ['Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)', 'Step_Time(s)']
    TARGET: str = 'Interpolated_SoH'
    SEQUENCE_LENGTH: int = 50
    LSTM_UNITS_1: int = 64
    LSTM_UNITS_2: int = 64
    DENSE_UNITS: int = 32
    DROPOUT_RATE: float = 0.2
    VALIDATION_SPLIT: float = 0.2
    EPOCHS: int = 50
    BATCH_SIZE: int = 64
    LEARNING_RATE: float = 0.001
    DATA_LOSS_WEIGHT: float = 1.0
    PHYSICS_LOSS_WEIGHT: float = 0.5
    USE_TUNER: bool = True # Set to True to enable hyperparameter tuning
    TUNER_MAX_TRIALS: int = 10
    TUNER_EPOCHS: int = 15

# --- Utility Classes ---
class DataLogger:
    """A simple class for styled and verbose logging."""
    @staticmethod
    def info(message: str) -> None: print(f"‚úÖ [INFO] {message}")
    @staticmethod
    def step(step_number: int, message: str) -> None: print(f"\n{'='*60}\n‚ñ∂Ô∏è  STEP {step_number}: {message}\n{'='*60}")
    @staticmethod
    def sub_step(message: str) -> None: print(f"\n--- {message} ---")
    @staticmethod
    def warning(message: str) -> None: print(f"‚ö†Ô∏è [WARNING] {message}")
    @staticmethod
    def result(message: str) -> None: print(f"üìä [RESULT] {message}")

# --- Core Components ---
class DataLoader:
    """Handles loading, splitting, standardizing, and sequencing of the entire dataset."""
    def __init__(self, config: Config):
        self.config = config
        self.data_path = os.path.join(config.DATA_PATH, config.REDUCED_DATA_FILE)
        self.full_df = self._load_data()
        self.scaler = StandardScaler()

    def _load_data(self) -> pd.DataFrame:
        """Loads the preprocessed data from a CSV file."""
        if not os.path.exists(self.data_path):
            raise FileNotFoundError(f"Data file not found at {self.data_path}")
        DataLogger.info(f"Loading data from {self.data_path}")
        return pd.read_csv(self.data_path)

    def get_cell_ids(self) -> List[str]:
        """Returns a list of unique cell IDs available in the dataset."""
        return self.full_df['cell_id'].unique().tolist()

    def _create_sequences(self, features: np.ndarray, targets: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
        """Creates sequences for LSTM training."""
        X, y = [], []
        for i in range(self.config.SEQUENCE_LENGTH, len(features)):
            X.append(features[i - self.config.SEQUENCE_LENGTH:i])
            y.append(targets[i])
        return np.array(X), np.array(y)

    def prepare_unified_data(self) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
        """Prepares training and validation data from the entire dataset."""
        DataLogger.sub_step("Preparing unified data for training and validation")

        train_df, val_df = train_test_split(
            self.full_df, test_size=self.config.VALIDATION_SPLIT, shuffle=False
        )

        DataLogger.info(f"Training data size: {len(train_df)} samples")
        DataLogger.info(f"Validation data size: {len(val_df)} samples")

        features_train = train_df[self.config.FEATURES].values
        self.scaler.fit(features_train)

        features_train_scaled = self.scaler.transform(features_train)
        features_val_scaled = self.scaler.transform(val_df[self.config.FEATURES].values)

        y_cols_for_loss = [self.config.TARGET, 'Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)']
        y_train_raw = train_df[y_cols_for_loss].values
        y_val_raw = val_df[y_cols_for_loss].values

        X_train, y_train = self._create_sequences(features_train_scaled, y_train_raw)
        X_val, y_val = self._create_sequences(features_val_scaled, y_val_raw)

        DataLogger.info(f"Train sequence shape: X={X_train.shape}, y={y_train.shape}")
        DataLogger.info(f"Validation sequence shape: X={X_val.shape}, y={y_val.shape}")

        return X_train, y_train, X_val, y_val

    def prepare_data_for_testing_cell(self, cell_id: str) -> Tuple[np.ndarray, np.ndarray]:
        """Prepares the full dataset for a single cell for testing, using the globally-fit scaler."""
        DataLogger.sub_step(f"Preparing test data for cell: {cell_id}")
        cell_df = self.full_df[self.full_df['cell_id'] == cell_id].copy()

        features_scaled = self.scaler.transform(cell_df[self.config.FEATURES].values)

        y_cols_for_loss = [self.config.TARGET, 'Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)']
        target_and_physics_data = cell_df[y_cols_for_loss].values

        X_test, y_test = self._create_sequences(features_scaled, target_and_physics_data)
        DataLogger.info(f"Test data shape for cell {cell_id}: X={X_test.shape}, y={y_test.shape}")
        return X_test, y_test

class PI_LSTMBuilder:
    """Builds the Physics-Informed LSTM model, accepting dynamic loss weights."""
    def __init__(self, config: Config, input_shape: Tuple):
        self.config = config
        self.input_shape = input_shape

    def _create_combined_loss(self, data_weight: float, phys_weight: float) -> callable:
        """Creates a combined loss function using the provided weights."""
        def combined_loss(y_true: tf.Tensor, y_pred: tf.Tensor) -> tf.Tensor:
            soh_true = y_true[:, 0]
            V_terminal_true = y_true[:, 1]
            I_true = y_true[:, 2]
            R0_true = y_true[:, 3]

            soh_pred = y_pred[:, 0]
            U_oc_pred = y_pred[:, 1]
            U_p1_pred = y_pred[:, 2]
            U_p2_pred = y_pred[:, 3]

            data_loss = tf.reduce_mean(tf.square(soh_true - soh_pred))
            V_terminal_pred = U_oc_pred - (I_true * R0_true) - U_p1_pred - U_p2_pred
            phys_loss = tf.reduce_mean(tf.square(V_terminal_true - V_terminal_pred))
            total_loss = (data_weight * data_loss) + (phys_weight * phys_loss)
            return total_loss
        return combined_loss

    def build_model(self, hp: kt.HyperParameters = None) -> keras.Model:
        """Builds the Keras model, handling hyperparameter-tuned loss weights."""
        lstm_units_1 = hp.Int('lstm_units_1', 32, 128, 32) if hp else self.config.LSTM_UNITS_1
        lstm_units_2 = hp.Int('lstm_units_2', 32, 128, 32) if hp else self.config.LSTM_UNITS_2
        dense_units = hp.Int('dense_units', 16, 64, 16) if hp else self.config.DENSE_UNITS
        learning_rate = hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4]) if hp else self.config.LEARNING_RATE
        dropout_rate = hp.Float('dropout_rate', 0.1, 0.5, 0.1) if hp else self.config.DROPOUT_RATE

        if hp:
            physics_loss_weight = hp.Float('physics_loss_weight', min_value=0.02, max_value=1.0, step=0.04)
            data_loss_weight = 1.0
        else:
            physics_loss_weight = self.config.PHYSICS_LOSS_WEIGHT
            data_loss_weight = self.config.DATA_LOSS_WEIGHT

        inputs = keras.Input(shape=self.input_shape)
        x = layers.LSTM(units=lstm_units_1, return_sequences=True)(inputs)
        x = layers.Dropout(dropout_rate)(x)
        x = layers.LSTM(units=lstm_units_2, return_sequences=False)(x)
        x = layers.Dropout(dropout_rate)(x)
        x = layers.Dense(units=dense_units, activation='relu')(x)
        x = layers.Dense(units=dense_units // 2, activation='relu')(x)
        outputs = layers.Dense(4, name='combined_output')(x)

        model = keras.Model(inputs=inputs, outputs=outputs)

        combined_loss_fn = self._create_combined_loss(data_loss_weight, physics_loss_weight)
        optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
        model.compile(optimizer=optimizer, loss=combined_loss_fn, metrics=['mae'])
        return model

class Trainer:
    """Orchestrates the unified model training process."""
    def __init__(self, config: Config, model_builder: PI_LSTMBuilder):
        self.config = config
        self.model_builder = model_builder

    def train_unified_model(self, X_train: np.ndarray, y_train: np.ndarray, X_val: np.ndarray, y_val: np.ndarray) -> str:
        """Trains a single unified model on the entire dataset."""
        DataLogger.sub_step("Training the Unified PI-LSTM Model")
        os.makedirs(self.config.MODEL_PATH, exist_ok=True)
        model_path = os.path.join(self.config.MODEL_PATH, self.config.UNIFIED_MODEL_NAME)

        if self.config.USE_TUNER:
            DataLogger.info("Starting hyperparameter tuning for the unified model...")
            tuner = kt.Hyperband(self.model_builder.build_model, objective='val_loss', max_epochs=self.config.TUNER_EPOCHS,
                                 factor=3, directory=os.path.join(self.config.MODEL_PATH, 'tuner'),
                                 project_name='unified_pi_lstm_tuner', overwrite=True)
            tuner.search(X_train, y_train, epochs=self.config.TUNER_EPOCHS, validation_data=(X_val, y_val))
            best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
            DataLogger.info(f"Best hyperparameters found: {best_hps.values}")
            model = self.model_builder.build_model(best_hps)
        else:
            DataLogger.info("Building unified model with manual hyperparameters...")
            model = self.model_builder.build_model()

        model.summary()
        checkpoint = keras.callbacks.ModelCheckpoint(model_path, monitor='val_loss', save_best_only=True, mode='min', verbose=1)
        early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, mode='min', restore_best_weights=True)

        DataLogger.info("Starting final model training...")
        history = model.fit(X_train, y_train, epochs=self.config.EPOCHS, batch_size=self.config.BATCH_SIZE,
                            validation_data=(X_val, y_val), callbacks=[checkpoint, early_stopping], verbose=1)

        self._plot_training_history(history)
        return model_path

    def _plot_training_history(self, history: keras.callbacks.History):
        plt.figure(figsize=(12, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Unified Model Training and Validation Loss')
        plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)
        plt.show()

class ModelEvaluator:
    """Handles loading the unified model and evaluating its performance on each individual cell."""
    def __init__(self, config: Config, data_loader: DataLoader):
        self.config = config
        self.data_loader = data_loader

    def evaluate_unified_model(self, model_path: str):
        """Loads the unified model and evaluates its performance on each cell individually."""
        DataLogger.sub_step(f"Evaluating unified model from {model_path} on all individual cells")

        dummy_builder = PI_LSTMBuilder(self.config, (self.config.SEQUENCE_LENGTH, len(self.config.FEATURES)))
        custom_loss = dummy_builder._create_combined_loss(
            data_weight=self.config.DATA_LOSS_WEIGHT,
            phys_weight=self.config.PHYSICS_LOSS_WEIGHT
        )

        try:
            model = keras.models.load_model(model_path, custom_objects={'combined_loss': custom_loss})
            DataLogger.info("Unified model loaded successfully.")
        except Exception as e:
            DataLogger.warning(f"Could not load the model from {model_path}. Error: {e}")
            return

        all_cell_ids = self.data_loader.get_cell_ids()
        for cell_id in all_cell_ids:
            try:
                X_test, y_test = self.data_loader.prepare_data_for_testing_cell(cell_id)

                if X_test.shape[0] == 0:
                    DataLogger.warning(f"Not enough data for cell {cell_id} to create test sequences. Skipping.")
                    continue

                DataLogger.info(f"Predicting SoH for test set of cell {cell_id}...")
                predictions = model.predict(X_test)

                y_true_soh = y_test[:, 0]
                y_pred_soh = predictions[:, 0]

                self._generate_report(y_true_soh, y_pred_soh, cell_id)
                self._plot_predictions(y_true_soh, y_pred_soh, cell_id)

            except Exception as e:
                DataLogger.warning(f"An error occurred while evaluating cell {cell_id}: {e}")
                import traceback
                traceback.print_exc()

    def _generate_report(self, y_true: np.ndarray, y_pred: np.ndarray, cell_id: str):
        mse = mean_squared_error(y_true, y_pred)
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        report_df = pd.DataFrame({
            'Metric': ['Mean Squared Error (MSE)', 'Mean Absolute Error (MAE)', 'R-squared (R¬≤)'],
            'Value': [f"{mse:.4f}", f"{mae:.4f}", f"{r2:.4f}"]
        })
        DataLogger.result(f"Evaluation Report for Cell {cell_id}:")
        print(report_df.to_string(index=False))

    def _plot_predictions(self, y_true: np.ndarray, y_pred: np.ndarray, cell_id: str):
        plt.figure(figsize=(15, 7))
        plt.plot(y_true, label='Actual SoH', color='blue', linewidth=2, alpha=0.8)
        plt.plot(y_pred, label='Predicted SoH', color='red', linestyle='--', linewidth=2, alpha=0.8)
        plt.title(f'Unified Model SoH Prediction vs. Actual for Cell: {cell_id}')
        plt.xlabel('Time Step (Sample)'); plt.ylabel('State of Health (SoH %)'); plt.legend(); plt.grid(True)
        plt.show()

# --- Main Pipeline Orchestrator ---
class TrainingPipeline:
    """Orchestrates the unified training and evaluation pipeline."""
    def __init__(self):
        self.config = Config()
        self.data_loader = DataLoader(self.config)

    def run(self):
        DataLogger.step(1, "Initializing Unified PI-LSTM Pipeline")
        os.makedirs(self.config.MODEL_PATH, exist_ok=True)

        DataLogger.step(2, "Preparing Unified Data")
        try:
            X_train, y_train, X_val, y_val = self.data_loader.prepare_unified_data()
            input_shape = (X_train.shape[1], X_train.shape[2])
        except Exception as e:
            DataLogger.warning(f"Failed to prepare data: {e}")
            return

        DataLogger.step(3, "Training Unified Model")
        try:
            builder = PI_LSTMBuilder(self.config, input_shape)
            trainer = Trainer(self.config, builder)
            trained_model_path = trainer.train_unified_model(X_train, y_train, X_val, y_val)
        except Exception as e:
            DataLogger.warning(f"An error occurred during training: {e}")
            import traceback
            traceback.print_exc()
            return

        DataLogger.step(4, "Post-Training Evaluation on Individual Cells")
        if not os.path.exists(trained_model_path):
            DataLogger.warning(f"Model file not found at {trained_model_path}. Skipping evaluation.")
            return

        evaluator = ModelEvaluator(self.config, self.data_loader)
        evaluator.evaluate_unified_model(trained_model_path)

        DataLogger.step(5, "PI-LSTM Pipeline Complete")

def main():
    try:
        from google.colab import drive
        print("üöÄ Mounting Google Drive...")
        drive.mount('/content/drive', force_remount=True)
    except ImportError:
        DataLogger.warning("Not in Colab. Ensure local paths are correct.")

    pipeline = TrainingPipeline()
    pipeline.run()

if __name__ == '__main__':
    main()
```
